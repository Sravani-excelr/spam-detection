```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.polarity_scores("The boy is crying for an icecream")
```

```python
# Create a new column for sentiment
df_final['Sentiment_0'] = df_final['text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df_final.head()
```

```python
df_final['Sentiment'] = df_final['Sentiment_0'].apply(lambda x: 'positive' if x > 0 else 'negative')
```

# Text classification

```python
df_final['text'] = df_final['text'].str.lower()
df_final.head()
```

```python
df_final['text']
```

```python
import string
df_final['text'] = df_final['text'].str.replace('[{}]'.format(string.punctuation), '')
df_final.head()
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
df_final['text'] = df_final['text'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))
df_final.head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df_final['text'] = df_final['text'].apply(lambda x: " ".join([word for word in x.split() if word not in stop_words]))
df_final.head()
```

```python
Y = df_final["Sentiment"]
x = df_final['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2),max_df=0.5,min_df=2,smooth_idf=True)
X = vectorizer.fit_transform(x)

x_new = X.toarray()

x_new = pd.DataFrame(x_new,columns=vectorizer.get_feature_names_out())
#x_new.to_csv("newdata.csv")
```

```python
x_new
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_new, Y)
```

```python
y_pred = model.predict(x_new)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y, y_pred)
import numpy as np
print("accuacy score: ",np.round(accuracy,2))
```
